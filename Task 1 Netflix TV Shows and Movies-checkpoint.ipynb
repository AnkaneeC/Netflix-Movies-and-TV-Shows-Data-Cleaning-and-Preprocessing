```python
df = pd.read_csv(r"D:\ELEVATE LABS\Task 1 21-04-2025\titles.csv")
df
```

```python
# Removing duplicates
```

```python
df = df.drop_duplicates()
df
```

```python
# Renaming columns with sae formatting
```

```python
df.columns # Checking column names
```

```python
df.columns = ['id', 'title', 'type', 'description', 'release_year',
       'age_certification', 'runtime (mins)', 'genres', 'production_countries',
       'seasons', 'imdb_id', 'imdb_score', 'imdb_votes', 'tmdb_popularity',
       'tmdb_score', 'date_added']
df
```

```python
df["title"] = df["title"].str.strip()  # Removing any leading or trailing spaces in film titles
df
```

```python
df.isnull().sum()  # Finding missing values
```

```python
df.info()
```

```python
df.loc[0:, "date_added"] = pd.to_datetime(df["date_added"])  # Updated date format
```

```python
df
```

```python
df = df.drop(columns = ['tmdb_popularity', 'tmdb_score']) #dropping columns
```

```python
df.head()
```

```python
for x in df.index:   # Removing data of shows/movies with an R rating
    if df.loc[x, "age_certification"] == 'R':
        df.drop(x, inplace = True)
df
```

```python
df.reset_index(drop = True)  # rearranging the index
```

```python
df = df.fillna('') # filling all NaN values with empty string
df
```

```python
df["type"] = df["type"].str.lower()  # Converting all text values in the column "type" to lowercase
df
```
